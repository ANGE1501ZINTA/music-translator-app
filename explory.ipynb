```python
result = model.transcribe('beyond-the-horizon-231945.mp3', fp16=False)
print(result["segments"])
```

```python
import re
```

```python
def clean_text(t):
    t = re.sub(r"\s+", " ", t)  # enlever espaces multiples
    t = t.strip()
    return t
```

```python
lines = [clean_text(seg["text"]) for seg in result["segments"] if seg["text"].strip()]
ignore_list = ["factory", "usine", "music"]
lines_cleaned = [line for line in lines if line.lower() not in ignore_list]

```

```python
from langdetect import detect
```

```python

full_text = " ".join(lines_cleaned )
language = detect(full_text)
print("Langue d√©tect√©e :", language)  
```

```python
from deep_translator import GoogleTranslator
```

```python
translated_lines = []
for line in lines_cleaned :
    translated_line = GoogleTranslator(source='auto', target='fr').translate(line)
    translated_lines.append(translated_line)
print(translated_lines[:300])
```

```python
with open("traduction_text.txt", "w", encoding="utf-8") as f:
    for t_line in translated_lines:
        f.write(t_line + "\n")  

print("Traduction texte enregistr√©e")
```
